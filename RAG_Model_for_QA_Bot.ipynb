```python
os.environ["PINECONE_API_KEY"] = "b2bb88c2-b9fa-4e21-a52e-7a28a36f81e5"

pc = Pinecone(api_key=os.environ.get("PINECONE_API_KEY"))

index_name = "financial-rag-qa"

# Create Pinecone

pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1")
    )

index = pc.Index(index_name)
```

```python
# Load embedding model
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Create documents by combining question and context
data = ds['train']
documents = [f"Question: {item['question']} Context: {item['context']}" for item in data]

# Embed documents and generate embeddings
def embed_documents_with_metadata(documents, model):
    embeddings = model.encode(documents, show_progress_bar=True)
    # Create tuples of (id, embedding, metadata)
    vectors = [(f'id_{i}', embedding, {'content': doc}) for i, (embedding, doc) in enumerate(zip(embeddings, documents))]
    return vectors

# Generate embeddings for the documents
vectors_with_metadata = embed_documents_with_metadata(documents, embedding_model)
```

```python
# Batch size for upserting vectors
batch_size = 100
for i in range(0, len(vectors_with_metadata), batch_size):
    batch_vectors = vectors_with_metadata[i:i + batch_size]
    index.upsert(vectors=batch_vectors)
```

```python
# Retrieve matching documents based on query
def retrieve_documents(query, index, model, top_k=1): # modify the k value
    # Embed the query
    query_embedding = model.encode(query)

    # Query Pinecone index
    results = index.query(
        vector=query_embedding.tolist(),
        top_k=top_k,
        include_metadata=True
    )

    # Extract matched documents' content from metadata
    matched_documents = [match['metadata']['content'] for match in results['matches']]

    return matched_documents
```

```python
import cohere

# Initialize Cohere client
cohere_api_key = 'TU6TQq3r8pHomFp7QTFjNrXVoNmR6ByAjfFYljtW'
co = cohere.Client(cohere_api_key)

# Generate response using retrieved documents
def generate_response(query):
    # Retrieve relevant documents from Pinecone
    documents = retrieve_documents(query, index, embedding_model)
    print("documents: ", documents)

    # Combine retrieved documents into a single context
    context = "\n".join(documents)
    print("context: ", context)

    # Create prompt for Cohere API
    prompt = f"{context}\n\nQ: {query}\nA:"
    print("prompt: ", prompt)

    # Generate response using Cohere
    response = co.generate(
        model='command-xlarge-nightly',
        prompt=prompt,
        max_tokens=100,
        temperature=0.7
    )
    return response.generations[0].text.strip()
```

```python
# Example user query
user_query = "What was the percentage change in revenue for the Compute & Networking segment from fiscal year 2022 to 2023?"
response = generate_response(user_query)
print("Response:", response)
```
